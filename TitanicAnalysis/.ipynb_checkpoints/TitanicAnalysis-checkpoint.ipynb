### Step2.数据处理

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import unicodecsv
from collections import defaultdict
sns.set(color_codes = True)

file = open('titanic-data.csv','rb')
data = pd.read_csv(file)
print data.info()
# print data 
# with open('titanic-data.csv','rb') as f:
#     reader = unicodecsv.DictReader(f)
#     data = list(reader)
```

**分析**：
PassengerId是乘客的编号不能为空，检查csv文件也能验证这一点，说明数据集中共有891名乘客。

Carbin这列数据只有204个有效值，而且编号很难找到规律，因此本次研究中暂不研究Carbin与生还率的关系；把Carbin中数据划分为1（有船舱号码）与0（缺失数据），方便观察其他列数据。

Age这一列存在数据缺失，有177人没有年龄数据（891-714），我猜测年龄大小与乘客船票等级有关，原因可能是年龄越大经济条件越好，能负担起较贵的船票；需要验证年龄与船票等级的关系，如果相关程度较大，可以用不同等级乘客年龄的平均值填充缺失的年龄数据。**同时**由于年龄数据范围比较大，直接观察不方便，因此要添加一列Age_Group将年龄分组。

Embarked这列有2个缺失值，找到并观察这条数据有何特征；Embarked这一列是字符，不方便计算相关系数，新建Aim_place列代替目的地：Q→1  C→2   S→3

Sex这一列全是字符，转换为数字更方便计算相关系数。新建gender列，其中女性为1男性为0。

```python
#目标1：修改船舱列，1为有船舱号码。0为数据缺失
def change_carbin(str):
    if str == 0:
        return 0
    else:
        return 1
data['Cabin'] = data['Cabin'].replace(np.nan,0)
data['Cabin'] = data['Cabin'].apply(change_carbin)
```

```python
#目标2：验证年龄与船票等级的关系，新建年龄分组列
pclass_age = data[['Pclass','Age']].dropna()
np.corrcoef(pclass_age['Pclass'],pclass_age['Age'])
```

**分析：** 相关系数r = -0.36922602 ,说明年龄与船票等级存在关系；r < 0、Pclass随着数值增大而等级降低，说明Age与Pclass是正相关，可以用不同组的平均值填充缺失的数据。

```python
#用每个船票级别的人的年龄去填充那些没有年龄的人 :直接修改NAN
age_mean = data.groupby('Pclass').sum()['Age'] / data.groupby('Pclass').size()  #series类型 123    
def fill_age(df):
    if pd.isnull(df['Age']):
        df['Age'] = age_mean[df['Pclass']]
    return df       
data['Age'] = data[['Age','Pclass']].apply(fill_age,axis=1)['Age']

#增加年龄分组列
bins = np.arange(0,90,10)
data['Age_Group'] = pd.cut(data['Age'],bins)
data.groupby('Age_Group')['PassengerId'].count()
```

```python
#目标：增加年龄的分组
# lim = 10
# def age_group(age,lim):
#     return '[%.f,%.f]'% ( lim * (age // lim),lim * (age // lim) + lim )

# age_group = pd.Series(age_group(age,lim) for  age in data['Age'])
# data['age_group'] = age_group
```

```python
#目标3：探索Embarked的缺失数据；新建Aim_place列
def print_Embarked_null(row):
    if pd.isnull(row['Embarked']):
        print row
s = data.apply(print_Embarked_null,axis=1)
```

**分析:** 可以看出这2位缺失目的地数据的女士在一个船舱，并且都获救了，都没有亲属关系，但她们没有明显特征，没有目的地可能是因为她们是船员家属或者是偷渡的或者记录数据时造成的错误，无法辨别。由于数量很少，在分析目的地与获救率时忽略这2个缺失。

```python
data['Aim_place'] = data['Embarked'].replace('Q',1.)
data['Aim_place'] = data['Aim_place'].replace('C',2.)
data['Aim_place'] = data['Aim_place'].replace('S',3.)
```

```python
#目的4：设置性别为数字：男性为0女性为1
data['Gender'] = data['Sex'].replace('male',0.)
data['Gender'] = data['Gender'].replace('female',1)
```

### Step3.探索数据集 

```python
%pylab inline
#画图：数量分布、与生还率关系、生还数量分布
def make_pictures(key):   
    data.groupby(key)['PassengerId'].count().plot(kind='pie',autopct='%0.f%%',figsize=(6,6))
    plt.title(key + ' VS Count')
    plt.show()
    
    data.groupby(key)['Survived'].mean().plot(kind='bar')
    plt.title(key + ' VS Servival Rate')
    plt.ylabel('Servival Rate')
    plt.show()
    
    data.groupby(['Survived',key])['Survived'].count().unstack().plot(kind='bar',stacked='True')
    plt.title(key + ' distribution in Survival ')
    plt.show()
#     key_servival_count = data.groupby(['Survived',key])['PassengerId'].count().plot(kind='pie')
#     plt.title('Servival & ' + key)
#     plt.show()
#计算相关系数R
def calcR(col1,col2):
    return np.corrcoef(data[col1],data[col2])
```

```python
#探索1：按照船票级别分组，查看不同级别的人获救百分比；画图然后计算相关性
# Survived_by_pClass = pClass.sum()/pClass.size()
# print Survived_by_pClass,pClass.size()
# Survived_by_pClass.plot()
make_pictures('Pclass')
calcR('Pclass','Survived')
```

**分析：**

由Pclass VS Count、Pclass VS Servival Rate 2张图可以看出，乘客最多的是第三等的客人，但他们生还率最低；第一、二等的客人数量相差不大，而第一等的客人生还率明显高于第二等。

由Pclass distribution in Survival可以看出，生还的乘客中船票是一等的人最多，丧生的乘客中三等乘客最多。

相关系数r=-0.33848104,而船票等级随数量增大而降低，说明Pclass与生还率呈正相关。

```python
#探索2：按性别分组，画柱状图看每个年龄段的生还率
make_pictures('Sex')
calcR('Gender','Survived')
```

**分析：**

由Sex VS Count和Sex VS Servival Rate两张图可以看出，乘客中女性只有男性一半但是女性生还率是男性的3倍多；由Sex distribution in Survival图可以看出，生还女性数量也是男性的2倍多。

计算性别与生还率的相关系数（女性为1男性为0）r = 0.54335138,也说明性别与生还率有明显关联，女性生还率更大。

```python
#探索3：计算每不同年龄段乘客的生还率
make_pictures('Age_Group')
calcR('Age','Survived')
```

**分析：**

由Age_Group VS Count可以看出10-50岁年龄段乘客是最多的，而在Age_Group VS Servival Rate图中，他们的生还率变化不大；在Age_Group distribution in Survival图中，不同年龄段生还数量与他们在Age_Group VS Count中总体人数的分布相似；

计算年龄与生还率的关系 r = -0.02955251比较小，初步判断年龄与生还率无关。

```python
#乘客不同目的地，他们的获救率有什么不同？
make_pictures('Embarked')
sa = data[['Survived','Aim_place']].dropna()
np.corrcoef(sa['Survived'],sa['Aim_place'])
```

**分析**

由Embarked VS Count、Embarked VS Servivl Rate图可以看出，去往S目的地的人最多但是生还率最低，去往C目的地的人较少但是生还率明显比其他人高；结合Embarked VS Count、Embarked distribution in Survival图可以看出，去往S目的地的人大约是去往C目的地人的4倍，但是生还的只是去往C的2倍；

计算不同目的地与生还率的相关系数 r=-0.10866867,说明去往不同目的地与生还率相关。结合三张图的展示，可以猜测去往C目的地的人生还率更高。

### Step4. 预测

#### 乘客生还率更高的乘客可能具备的特征包括：
        1.船票等级高
        2.女性
        3.目的地是C
#### 预测结果中可能存在的限制：

1.在数据处理阶段使用乘客船票等级对应地年龄的平均值去填充没有年龄的乘客，可能造成预测结果不准确；

2.乘客的目的地不同，可能是因为他们来自不同地方，从而造成他们的水性不同或者身体素质不同，因此可能是预测乘客生还率的一个条件，但是并不绝对；计算Embarked时用数字指代目的地，数字大小可能影响了相关系数的计算；

3.获救的乘客中女性占很大部分，可能是先让女士小孩老人上救生船造成的——这可能是在绝大多数乘客道德优秀的条件下的表现,但是如果道德品质差或者没有强有力的道德保证，会造成什么样的后果就不确定；

4.乘客的生还率可能还与救生船的数量、逃生时带的食物的数量等有关，但是并没有给出相关数据。
